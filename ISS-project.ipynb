```python
import numpy as np
import matplotlib.pyplot as plt
import IPython
import math
from numpy.core.numeric import isclose
from scipy.signal import spectrogram, lfilter, freqz, find_peaks, butter
from scipy.io import wavfile
```

# Natanie zkladnho signlu
Zo vstupnho sboru 'xondru18.wav' natavam audio nahrvku. <br>
Do premennej fs je uloen vzorkovacia frekvencia, v mojom prpade sa jedn o 16kHz.<br>
Obsah audio nahrvky je uloen do premennej data.
<br>
as zistm pomocou nasledujceho vzorca:
\begin{equation*}
as = \frac{poet \: vzorkov}{vzorkovacia \: frekvencia}
\end{equation*}

```python
fs, data = wavfile.read("xondru18.wav")
t = np.arange(data.size) / fs
plt.figure(figsize=(15, 6))
plt.plot(t, data)
plt.xlabel("t[s]")
plt.ylabel("Amplituda")
plt.title("Zvukovy signal")
IPython.display.display(IPython.display.Audio(data, rate=fs))
```

```python
print("Minimalna hodnota :", data.min())
print("Maximalna hodnota :", data.max())
print("Dlzka nahravky(vo vzorkoch) :", len(data))
print("Dlzka nahravky(v sekundach) :", len(data) / fs, "s")
print("Vzorkovacia frekvencia :", fs, "Hz")
plt.figure(figsize=(15, 6))
plt.plot(t, data)
plt.xlabel("t[s]")
plt.ylabel("Amplituda")
plt.title("Normalizovany zvukovy signal")
```

# Usmernenie - Odstrnenie strednej hodnoty
Stredn hodnotu odstrnim odtanm strednej hodnoty pre kad vzorku signlu

```python
mean = np.mean(data)
for n in range(0, data.size):
    data[n] = data[n] - mean
```

# Normalizcia
Signl je potreba pred alm krokom normalizova do dynamickho rozsahu -1 a 1. <br>
Signl normalizujem pomocou vzorca:
\begin{equation*}
data = \frac{data}{max(\;data.max\;,\;abs(data.min)\;)}
\end{equation*}

```python
data = data / max(data.max(), -data.min())
refdata = data
plt.figure(figsize=(15, 6))
plt.grid(alpha=0.5, linestyle='--')
plt.title("Normalizovany signal")
plt.xlabel("t(s)")
plt.ylabel("Amplituda")
plt.plot(t, data)
```

# Rmcovanie
Kee celkov signl m dku 27034 vzorkov budem na spracovanie potrebova rmce, ktor maj by dlh 1024 vzorkov a maj ma prekrytie 512 vzorkov.<br>
Poet rmcov viem zisti zo vzorca:
\begin{equation*}
Poet\;rmcov = 1+\frac{len(data)-len(ramec)}{len(prekrytie)}
\end{equation*}


```python
n = 1024  # velkost ramca
m = 512  # dlzka prekrytia
array = [data[i : i + n] for i in range(0, len(data), n - m)]
index = 20  # index matice na vykreslenie
plt.figure(figsize=(15, 6))
plt.grid(alpha=0.5, linestyle='--')
plt.plot(np.arange(index * 512, index * 512 + 1024) / fs, array[index])
plt.title("Ramec signalu")
plt.xlabel("t(s)")
plt.ylabel("Amplituda")
```

# DFT
## Implementcia vlastnej Diskrtnej Fourierovej Transformcie
Pomocou vlastnej implementcie DFT vykreslm frekvencie z rmca 20.

```python
def dft(array):
    array = np.append(array, np.zeros(1024 - len(array)))
    arr = []
    for i in range(len(array)):
        coef = 0
        for j in range(len(array)):
            coef += array[j] * np.exp(-2j * np.pi * i * j * (1 / len(array)))
        arr.append(coef)
    return arr
#     N = len(array)
#     n = np.arange(N)
#     k = n.reshape((N, 1))
#     e = np.exp(-2j * np.pi * k * n / N)
#     X = np.dot(e, array)
#     return X
```

```python
index = 20
after_dft = dft(array[index])
y = np.arange(0, 1024 / 2) * (fs / 2048 * 2)  # pre 512 vzorkov rozsah od 0Hz do 8kHz
plt.figure(figsize=(15, 6))
plt.grid(alpha=0.5, linestyle='--')
plt.plot(y, np.abs(after_dft[:512]))
plt.title("DFT")
plt.xlabel("f(Hz)")
plt.ylabel("Amplituda")
```

# Kontrola DFT
Pre kontrolu sprvnosti mojej DFT si vykreslm indentick rmec pomocou kninicovej funkcie **np.fft.fft()**<br>
Tie teraz porovnm funkciou **np.allclose()**, ktor vrti *true* ak s si podobn a *false* ak sa nepodobaj.<br>

```python
correctDFT = np.fft.fft(array[index])
if np.allclose(after_dft, correctDFT):
    print("DFTs match")
else:
    print("Something is wrong")
```

# Spektogram
Pre generovanie vlastnho spektogramu som postupne cez kad rmec robil naslednovn veci:<br>
- ***Predspracovanie*** aktulneho rmca obsahom tak, aby przdne polia boli vyplnen nulami
- ***DFT*** aktulneho rmca
- ***Odseknutie*** polovice rmca, kee po DFT je rmec symetrick a ja potrebujem iba 1/2
- ***P[k] = 10 log10 |X[k]|^2*** na upravenie hodnt aktulneho rmca do "logaritmickho vkonovho spektogramu"
- ***Uloenie*** vslednho rmca do poa

```python
frames = array
spectrum = []

for i in range(53): # jednotliv rmce
    frames[i] = np.pad(frames[i], (0, 1024 - len(frames[i])), 'constant')
    dftted = np.fft.fft(frames[i])
    spectrum.append(10 * np.log10(np.abs(dftted[:512])**2))
```

Pre korektn vyzobrazenie dt je potreba otoi vsledn dta o 90 pomocou funkcie **np.rot90()**

```python
plt.figure(figsize=(15,6))
plt.imshow(np.rot90(spectrum), extent=[0, 1.7, 0, 8000], aspect='auto')
plt.gca().set_title('Spektogram signalu')
plt.gca().set_xlabel('Cas [s]')
plt.gca().set_ylabel('Frekvencia [Hz]')
cbar = plt.colorbar()
cbar.set_label('Spektralna hustota vkonu [dB]', rotation=270, labelpad=15)
 
plt.tight_layout()
```

# Zistenie ruivch frekvenci
Ruiv frekvencie zskavam pomocou funkcie **find_peaks()**, ktor had vrcholy cez cel rmec.<br>
*Height=1* hovor, e funkcia bude vraca hodnoty, ktor s vaie ako 1.<br>
Aby som vedel priradi vrcholu frekvenciu, prepotam vrcholy pomocou vzorcu:
    \begin{equation*}
Peak = Peak*\frac{fs}{1024}
\end{equation*}
alej tmto vrcholom odstrnim desatinn as a nsledne ich vypem.

```python
dftSpctr = np.fft.fft(array[1])
after_dft_abs = np.abs(dftSpctr)
peaks, _ = find_peaks(after_dft_abs[: len(after_dft) // 2], height=1)
peak_vals = np.abs(dftSpctr[peaks])
peaks = peaks * (fs / 1024)
peaks = [np.floor(float(x)) for x in peaks]
print("Rusive frekvencie : ", peaks)
```

# Kontrola zistench ruivch frekvenci
Porovnvam, i s ruiv frekvencie s harmonicky zdruen (s nsobnky prvej frekvencie).<br>
Ak je frekvencia harmonicky zdruen, vypisujem e je OK

```python
peak = peaks[0]
before_peak = peaks[0]
diff = 15
for i in peaks:
    if math.isclose(peak, i, abs_tol=diff):
        print("Peak", i, "is OK")
    else:
        print("Peak", i, "is NOT OK")
    peak += before_peak
    diff += 15
```

# Generovanie signlu
Pre kontrolu ruivho signlu posluchom si potrebujem vygenerova signl zo tyroch ruivch frekvenci, tie vygenerujem tak, e pre kad vzorok signlu vypotam set tyroch kosnusoviek, ktor nsledne zapem do vslednho signlu. 

```python
k = len(data) / fs
time = np.arange(0, len(data), 1)
arr = []
for i in range(0, len(time)):
    item = 0
    for y in range(0, 4):
        item += peak_vals[y] * np.cos(2 * np.pi * k * i / len(time) * peaks[y])
    arr.append(item)
fig, ax = plt.subplots(2, 1, figsize=(15, 12))
ax[0].grid(alpha=0.5, linestyle='--')
ax[1].grid(alpha=0.5, linestyle='--')
ax[0].plot(time, arr)
ax[0].set_title("Generovany signal")
ax[0].set_xlabel("t(s)")
ax[0].set_ylabel("Amplituda")
ax[1].plot(time[:512], arr[:512])
ax[1].set_title("Usek generovaneho signalu (prvych 512 vzorkov)")
ax[1].set_xlabel("t(s)")
ax[1].set_ylabel("Amplituda") 
```

# DFT vygenerovaneho signlu
Pre kontrolu vygenerovanho signlu si ho nechm zobrazi cez DFT, aby som videl e sa v om nachdzaj len chcen frekvencie.<br> Vizulne porovnm s rmcom v ktorom som hadal ruiv frekvencie.

```python
correctDFT = np.fft.fft(arr[:1024])
_, ax = plt.subplots(2,1,figsize=(15, 12))
y = np.arange(0, 1024 / 2) * (fs / 2048 * 2)  # pre 512 vzorkov rozsah od 0Hz do 8kHz
ax[0].plot(y, np.abs(correctDFT[:512]))
ax[0].set_xlabel('f(Hz)')
ax[0].set_title('DFT vygenerovaneho signalu')
ax[0].grid(alpha=0.5, linestyle='--')
correctDFT = np.fft.fft(array[1])
ax[1].plot(y, np.abs(correctDFT[:512]))
ax[1].set_xlabel('f(Hz)')
ax[1].set_title('DFT mojho signalu')
ax[1].grid(alpha=0.5, linestyle='--')
plt.tight_layout()
```

# Spektogram vygenerovanho signlu
Takisto ako DFT tak si vygenerovan signl mem skontrolova aj cez spektogram, v ktorom vidm, e moje ruiv frekvencie svietia vrazne ltou farbou a vidm, e ich frekvencie sa rovnaj mojm ruivm.

```python
arr = np.asarray(arr, dtype=np.float64)
f, t, sgr = spectrogram(arr, fs)
sgr_log = 10 * np.log10(abs(sgr + 1e-20) ** 2)
plt.figure(figsize=(15, 6))
plt.pcolormesh(t, f, sgr_log)
plt.gca().set_xlabel("Cas [s]")
plt.gca().set_ylabel("Frekvencia [Hz]")
cbar = plt.colorbar()
cbar.set_label("Spektralna hustota vykonu [dB]", rotation=270, labelpad=15)
```

# Filtrovanie signlu
Na filtrovanie signlu si potrebujem navrhn sadu filtrov, ktor bud vedie potla tieto frekvencie.<br>
Vybral som si metdu 3. **nvrh 4 psmovch zdr**.<br>
Tieto zdre budem generova funkciou **butter()**<br>
Fukcia je volan s nasledujcimi parametrami:<br>
- ***Rd*** som nastavil na tret, lebo som sa rozhodol, e nm vygenerovan filtre vyzerali najlepie.<br>
- ***Medzn frekvencie*** tie s zapsan nasledovne \[ peaks(i) - 50 , peaks(i) + 50 \].<br>
- ***Typ filtra*** ten som nastavil na 'bandstop', aby filter vyzeral pribline takto ''''''''\\/'''''''.<br><br>
Nasledovne tmito filtrami vyfiltrujem signl.<br>
Pre vizualizciu fitrov ich ukem na grafe.

```python
order = 3
fig, ax = plt.subplots(2, 1, figsize=(15, 12))
offset = 50 # sirka 50Hz z kazdej strany
for i in peaks:
    lowcut = i - offset
    highcut = i + offset
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype="bandstop")
    data = lfilter(b, a, data)
    freq, h = freqz(b, a, fs=fs)
    ax[0].plot(freq[:256], 20 * np.log10(abs(h[:256])))
    ax[0].set_title('Modul frekvencej charakteristiky $|H(e^{j\omega})|$')
    ax[0].set_xlabel('Frekvencia [Hz]')
    ax[1].plot(freq[:256], np.unwrap(np.angle(h[:256])))
    ax[1].set_title('Argument frekvencnej charakteristiky $\mathrm{arg}\ H(e^{j\omega})$')
    ax[1].set_xlabel('Frekvencia [Hz]')
    



# ODDELENE
# order = 3
# offset = 50 # sirka 50Hz z kazdej strany
# for i in peaks:
#     lowcut = i - offset
#     highcut = i + offset
#     nyq = 0.5 * fs
#     low = lowcut / nyq
#     high = highcut / nyq
#     b, a = butter(order, [low, high], btype="bandstop")
#     data = lfilter(b, a, data)
#     freq, h = freqz(b, a, fs=fs)
   
#     _, ax = plt.subplots(1, 2, figsize=(15,6))

#     ax[0].plot(freq / 2 / np.pi * fs, np.abs(h))
#     ax[0].set_xlabel('Frekvencia [Hz]')
#     ax[0].set_title('Modul frekvencej charakteristiky $|H(e^{j\omega})|$')
#     ax[0].grid()    
#     ax[1].plot(freq / 2 / np.pi * fs, np.angle(h))
#     ax[1].set_xlabel('Frekvencia [Hz]')
#     ax[1].set_title('Argument frekvencnej charakteristiky $\mathrm{arg}\ H(e^{j\omega})$')
#     ax[1].grid()
    for ax1 in ax:
        ax1.grid(alpha=0.5, linestyle='--')

    plt.tight_layout()
```

# Nuly a Ply
Vypotam nulov body a ply mojich navrhnutch filtrov pomocou funkcie **butter s parametrom output='zpk'**, ktor vracia nuly a ply.<br>Tie zobrazujem v komplexnej rovine.

```python
order = 3
pointer = 1
offset = 50 # sirka 50Hz z kazdej strany
for i in peaks:
    lowcut = i - offset
    highcut = i + offset
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    z,p,k = butter(order, [low, high], btype="bandstop", output='zpk')
    plt.figure(figsize=(8, 8))
    ang = np.linspace(0, 2*np.pi,100)
    plt.plot(np.cos(ang), np.sin(ang))
    plt.title("Nuly a poly pre frekvenciu:"+str(i)+"Hz pre pasmovu zadrz")
    plt.scatter(np.real(z), np.imag(z), marker='o', facecolors='none', edgecolors='r', label='nuly')
    plt.scatter(np.real(p), np.imag(p), marker='x', color='g', label='ply')
    plt.gca().set_xlabel('Realn sloka $\mathbb{R}\{$z$\}$')
    plt.gca().set_ylabel('Imaginarn sloka $\mathbb{I}\{$z$\}$')
    plt.grid(alpha=0.5, linestyle='--')
    plt.legend(loc='upper right')
    plt.tight_layout()
    pointer += 1
```

# Frekvenn charakteristika
Pre zistenie frekvennej charakteristiky filtra som vytvoril signl dlh 100 vzorkov ktor je nulov, okrem prvho vzorku, ktor je 1. Cez tento signl preeniem postupne filtre a sledujem, o sa zo signlom bude dia.

```python
order = 3
offset = 50 # sirka 50Hz z kazdej strany
for i in peaks:
    lowcut = i - offset
    highcut = i + offset
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype="bandstop")
    print("Pre frekvenciu:",i,"su koeficienty A a B:")
    np.set_printoptions(precision=5)
    print("A: ",a)
    print("B: ",b)
    N_imp = 100 # 100 vzorkov
    imp = [1, *np.zeros(N_imp-1)] # inicializacia vzorkov
    h = lfilter(b, a, imp) # filtracia
    plt.figure(figsize=(15, 6))
    plt.stem(np.arange(N_imp), h, basefmt=' ')
    plt.gca().set_xlabel('$n$')
    plt.gca().set_title('Impulzna odozva $h[n]$ pre frekvenciu:'+str(i)+"Hz")
    plt.grid(alpha=0.5, linestyle='--')
    plt.tight_layout()
```
